```python
import pandas as pd
import pyarrow.parquet as pq
import os

# Define the paths to the original Parquet files
parquet_paths = [
    r"C:\Users\reece\Downloads\Final Project - Career Foundry\Final working files\Monthly Borough Taxi Reports\Orignal Data\Original Parquet\yellow_tripdata_2023-04.parquet",
    r"C:\Users\reece\Downloads\Final Project - Career Foundry\Final working files\Monthly Borough Taxi Reports\Orignal Data\Original Parquet\yellow_tripdata_2023-03.parquet",
    r"C:\Users\reece\Downloads\Final Project - Career Foundry\Final working files\Monthly Borough Taxi Reports\Orignal Data\Original Parquet\yellow_tripdata_2023-02.parquet",
    r"C:\Users\reece\Downloads\Final Project - Career Foundry\Final working files\Monthly Borough Taxi Reports\Orignal Data\Original Parquet\yellow_tripdata_2023-01.parquet"
]

# Define the path to the CSV export folder
csv_folder = r"C:\Users\reece\Downloads\Final Project - Career Foundry\Final working files\Monthly Borough Taxi Reports\Orignal Data\Orignal CSV"

# Create the CSV folder if it doesn't exist
os.makedirs(csv_folder, exist_ok=True)

# Convert Parquet files to CSV and store in the CSV folder
for parquet_path in parquet_paths:
    # Read the Parquet file
    table = pq.read_table(parquet_path)
    df = table.to_pandas()

    # Get the filename without extension
    filename = os.path.splitext(os.path.basename(parquet_path))[0]

    # Define the CSV file path
    csv_path = os.path.join(csv_folder, f"{filename}.csv")

    # Export the DataFrame to CSV
    df.to_csv(csv_path, index=False)

    print(f"Converted {parquet_path} to {csv_path}")
```

```python
#Pick Up Time Flag
import datetime

# Function to map pickup times to time of day categories
def map_time_of_day(pickup_time):
    datetime_obj = datetime.datetime.strptime(pickup_time, '%Y-%m-%d %H:%M:%S')
    if 0 <= datetime_obj.hour < 6:
        return '12:01am - 6am, Early Morning'
    elif 6 <= datetime_obj.hour < 12:
        return '6:01am - 12pm, Morning'
    elif 12 <= datetime_obj.hour < 18:
        return '12:01pm - 6pm, Mid-Afternoon'
    elif 18 <= datetime_obj.hour < 24:
        return '6:01pm - 12am, Evening'
    else:
        return 'Unknown'

# Create the "PU_time_of_day" column based on "tpep_pickup_datetime"
combined_df['PU_time_of_day'] = combined_df['tpep_pickup_datetime'].apply(lambda x: map_time_of_day(x))


```

```python
#3.3 create borough flag

#Importing taxi zone data so we can convert the numerical values in Pick up location and drop off location ID

# Create file paths for the CSV files
path7 = r"C:\Users\reece\Downloads\Final Project - Career Foundry\taxi+_zone_lookup.csv"
taxi_zones_path = os.path.join(path7)
# Read the taxi zones CSV file into a DataFrame
taxi_zones_df = pd.read_csv(taxi_zones_path)
taxi_zones_df.head()
```

```python
# This command takes Location, Zone, and Burough data from taxi_zones and match's it with the PU Location Column
#I will need to do the same thing for the Drop Off column 
# Select the desired columns from taxi_zones_df
taxi_zones_subset = taxi_zones_df[['LocationID', 'Borough', 'Zone']]

# Merge the subset DataFrame into combined_df
combined_df = pd.merge(taxi_zones_subset, combined_df, left_on='LocationID', right_on='PULocationID', how='left')

# Rename the columns in combined_df
combined_df = combined_df.rename(columns={'Borough': 'Pick Up Borough', 'Zone': 'Pick Up Zone'})

combined_df.head()

```

```python
# Select the desired columns from taxi_zones_df for Drop Off
taxi_zones_subset_dropoff = taxi_zones_df[['LocationID', 'Borough', 'Zone']]

# Merge the Drop Off subset DataFrame into merged_df
combined_df = pd.merge(combined_df, taxi_zones_subset_dropoff, left_on='DOLocationID', right_on='LocationID', how='left')

# Rename the columns in merged_df for Drop Off
combined_df = combined_df.rename(columns={'Borough': 'Drop Off Borough', 'Zone': 'Drop Off Zone'})


combined_df.head()
```
